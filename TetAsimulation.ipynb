```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

There are a couple ways to parameterize these equations. For instance, I can use multiplicative Hill terms to capture how growth is suppressed based on internal Tet concentration $T_{in}$, as well as toxicity effects based on enzyme activity (assumed proportional to gene copy number) $\rho$.

Generically:  

$r = r_{max} * \frac{K_1^{n_1}}{K_1^{n_1} + \rho^{n_1}} * \frac{K_2^{n_2}}{K_2^{n_2} + T_{in}^{n_2}}$

By letting $r_{max} = K_2 = n_1 = n_2 = 1$,  
calling $K_1$ $\rho_{threshold}$  
and rearranging, we get:  
$r = \frac{1}{(\rho/\rho_{threshold} + 1)} * \frac{1}{1+T_{in}}$

This last equation is equation S19 from the 2013 antibiotic resistance landscape bistability Science paper with a new multiplicative cost term based on discussions with Emrah and Jia.

```python
def solve_Tin_no_cost(Tex, rho, sigma):
    # Define the equation S16
    equation = lambda Tin: Tex - Tin * (1 + 1 / (1 + Tin) * rho / (1 + sigma * Tin))

    # Solve for Tin
    supposed_solutions = fsolve(equation, [0.00000001, 1])

    # Check the solutions for correctness.
    solutions = []
    solution_tolerance = 0.2
    for _Tin_ in supposed_solutions:
        check = Tex - _Tin_ * (1 + 1 / (1 + _Tin_) * rho / (1 + sigma *_Tin_))
        if abs(check) < solution_tolerance:
            solutions.append(_Tin_)
    
    # Keep only the positive solutions. negative solutions should be close to zero, so set to zero.
    positive_solutions = [max(x,0) for x in solutions]

    # Return the solutions based on the number of positive solutions
    if len(positive_solutions) == 0:
        ## let Tin = Tex in this case.
        Tin_low = Tin_high = Tex
    elif len(positive_solutions) == 1: #only one solution, so Tin_low = Tin_high.
        Tin_low = Tin_high = positive_solutions[0]
    else: ## this is a cubic equation so return the last value.
        Tin_low = positive_solutions[0]
        Tin_high = positive_solutions[-1]
    return Tin_low, Tin_high
```

```python
def solve_Tin_with_cost(Tex, rho, sigma, rho_threshold):
    # Define the modified equation S16 with multiplicative cost term.
    equation = lambda Tin: Tex - Tin * (1 + (1 / (1 + Tin) * 1/(rho/rho_threshold + 1) * rho / (1 + sigma * Tin)))

    # Solve for Tin
    supposed_solutions = fsolve(equation, [0.00000001, 1])

    # Check the solutions for correctness.
    solutions = []
    solution_tolerance = 0.2
    for _Tin_ in supposed_solutions:
        check = Tex - _Tin_ * (1 + (1 / (1 + _Tin_) * 1/(rho/rho_threshold + 1) * rho / (1 + sigma * _Tin_)))
        if abs(check) < solution_tolerance:
            solutions.append(_Tin_)

    # Keep only the positive solutions. negative solutions should be close to zero, so set to zero.
    positive_solutions = [max(x,0) for x in solutions]

    # Return the solutions based on the number of positive solutions
    if len(positive_solutions) == 0:
        ## let Tin = Tex in this case.
        Tin_low = Tin_high = Tex
    elif len(positive_solutions) == 1: #only one solution, so Tin_low = Tin_high.
        Tin_low = Tin_high = positive_solutions[0]
    else: ## this is a cubic equation so return the last value.
        Tin_low = positive_solutions[0]
        Tin_high = positive_solutions[-1]
    return Tin_low, Tin_high
```

```python
def logistic_growth_no_cost(t, N, rho, Tex, sigma):
    
    if N < 0: N = 0 ## ensure population is non-negative
        
    Tin_low, Tin_high = solve_Tin_no_cost(Tex, rho, sigma) ## get the stable solutions

    #print(Tin_low, Tin_high)

    if N > 0.1: ## if the population is large enough, then there is a subpopulation of fast-growers (Tin_low).
        Tin = Tin_low
    else:
        Tin = Tin_high
    
    r = (1 / (1 + Tin))
    dNdt = r * N * (1 - N)
    return dNdt
```

```python
def logistic_growth_with_cost(t, N, rho, Tex, sigma, rho_threshold):
    
    if N < 0: N = 0 ## ensure population is non-negative
        
    Tin_low, Tin_high = solve_Tin_with_cost(Tex, rho, sigma, rho_threshold) ## get the stable solutions

    #print(Tin_low, Tin_high)

    if N > 0.1: ## if the population is large enough, then there is a subpopulation of fast-growers (Tin_low).
        Tin = Tin_low
    else:
        Tin = Tin_high
    
    r = 1 / (1 + Tin) * 1/(rho/rho_threshold + 1)
    dNdt = r * N * (1 - N)
    return dNdt
```

```python
# Parameters
N0 = 0.01  # Initial population size

# E0 is Initial enzyme concentration (rho)
#E0 = 0 # no enzyme
#E0 = 180 # low copy number (no plasmid)
E0 = 1000 # high copy number (pUC plasmid)

Tex = 30  # External antibiotic
sigma = 1  # default enzyme parameter that is fixed.
rho_threshold = 500
```

```python
# Time range for integration
t_span = (0, 500)

# Solve the ODE
solution = solve_ivp(logistic_growth_with_cost, t_span, [N0], args=(E0, Tex, sigma, rho_threshold), dense_output=True)

# Time points for evaluation
t_eval = np.linspace(0, 500, 100)

# Evaluate the solution at the desired time points
population = solution.sol(t_eval)

# Plotting the results
plt.plot(t_eval, population[0])
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.title('Logistic Growth with E, Tin, and their Relationship')
plt.grid(True)
plt.show()
```

```python
# Combinations of E and Tex

# E = 0, no TetA expression
# E = 180 TetA expressed from chromosome
# E = 1000, TetA expressed from pUC

combinations = [
    (0, 0, "E = 0, Tex = 0"),
    (0, 30, "E = 0, Tex = 30"),
    (180, 0, "E = 180, Tex = 0"),
    (180, 30, "E = 180, Tex = 30"),
    (1000, 0, "E = 1000, Tex = 0"),
    (1000, 30, "E = 1000, Tex = 30")
]

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

for E, Tex, label in combinations:
    # Solve the ODE
    solution = solve_ivp(logistic_growth_with_cost, t_span, [N0], args=(E, Tex, sigma, rho_threshold), dense_output=True)
    population = solution.sol(t_eval)

    # Select the corresponding axis based on Tex
    if Tex == 0:
        ax = axs[0]
    elif Tex == 30:
        ax = axs[1]

    # Plot the population curve
    ax.plot(t_eval, population[0], label=label)

# Set the labels and titles for each panel
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Population Size')
axs[0].set_title('Tex = 0')
axs[0].grid(True)

axs[1].set_xlabel('Time')
axs[1].set_ylabel('Population Size')
axs[1].set_title('Tex = 30')
axs[1].grid(True)

# Add a legend
axs[0].legend(loc='lower right')
axs[1].legend(loc='lower right')

# Adjust spacing between subplots
plt.tight_layout()

## Save this plot as Figure 1F.
plt.savefig('../results/Fig1F.pdf', format='pdf')

# Show the plot
plt.show()
```

```python
## for debugging, let's plot modified equation S19 with multiplicative cost term. This is what matters: 

# Define the equation as a lambda function.
# IMPORTANT: note that this is "x" and not "Tex - x" as in the solver above.
S19_equation_for_growth_rate_with_cost_term = lambda Tin: (1 / (1 + Tin)) * (1 / (E0 / rho_threshold + 1))

# Generate values for Tin
Tin_values = np.linspace(0, 100, 100)  # Adjust the range if needed

# Calculate the corresponding values for the equation
result_values = S19_equation_for_growth_rate_with_cost_term(Tin_values)

# Plot the equation
plt.plot(Tin_values, result_values)
plt.xlabel('Tin')
plt.ylabel('growth rate')
plt.title('Growth rate from Tin based on modified equation S19 with cost term')
plt.grid(True)
plt.show()
```
